```python
N = 100
line = hpw.Line(N)
line 
```

```python
qw = hpw.Coined(line)
qw 
```

```python
vertex = (2*N-2) // 2
# initial_state = qw.ket(vertex, vertex + 1)

initial_state = np.zeros(2*N-2,dtype=complex)
initial_state[vertex] = 1/np.sqrt(2)
initial_state[vertex+1] =(-1j)*(1/np.sqrt(2))
```

```python
final_state = qw.simulate(time=50,
                          initial_state=initial_state)
```

```python
probability = qw.probability_distribution(final_state)
# type(probability)
# print(probability)
```
